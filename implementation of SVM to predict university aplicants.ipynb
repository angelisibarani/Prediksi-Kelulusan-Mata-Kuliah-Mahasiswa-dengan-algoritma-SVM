# Data Preparation

```python
from sklearn.preprocessing import StandardScaler 

X = df.drop(["ceeb_id", "fullname", "campus", "city.x", "fall_term", "school_name", "ethcat", "col_lat", "col_long", "sch_lat", "sch_long", "d"], axis = 1)
X = pd.get_dummies(X, drop_first=True)
scaler = StandardScaler()
cols = X.columns
X = pd.DataFrame(scaler.fit_transform(X))
X.columns = cols
y = df["col_rank"]
```

```python
df.head()
```

# Modelling & Evaluation

```python
from sklearn.svm import LinearSVC
from sklearn.model_selection import KFold
import os
```

```python
scores = []
kfold = KFold(n_splits=5, random_state = 42, shuffle=True)
for train_index, test_index in kfold.split(X):
    model = LinearSVC()
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y[train_index], y[test_index]
    model.fit(X_train, y_train);
    scores.append(model.score(X_test, y_test))
    
```

```python
result = sum(scores)/len(scores)
```

```python
print(result)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC  
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
```

```python
# Create a pipeline
pipe = Pipeline([('classifier', LogisticRegression())])
```

```python
# Create space of candidate learning algorithms and their hyperparameters
search_space = [{'classifier': [LogisticRegression()]},
        {'classifier': [SVC()]},
        {'classifier': [KNeighborsClassifier(n_neighbors=9)]}] #9 ranks of colleges
```

```python
# Create grid search 
clf = GridSearchCV(pipe, search_space)
```

```python
# Fit grid search
best_model = clf.fit(X_train, y_train)
```

```python
# View best model
print best_model.best_estimator_.get_params()['classifier'], "\n"
print "Accuracy of our best model is", clf.score(X_test, y_test)*100, "%",  "\n"
print "Classification Report:", "\n", classification_report(y_test, best_model.predict(X_test))
```
